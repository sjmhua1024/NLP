```python
import nltk
from collections import Counter
import numpy as np
```

```python
with open('Poetry.txt',encoding='utf-8') as f:
    lines = f.readlines()
    wordFreqDict = dict()
    for line in lines:
        tokens = dict(nltk.FreqDist(list(line)))
        wordFreqDict = dict(Counter(wordFreqDict) + Counter(tokens))
    wordFreqTuple = sorted(wordFreqDict.items(),key = lambda x: x[1], reverse=True)
    fw = open('vocab.txt','w',encoding='utf-8')
    vocab = wordFreqTuple[:8000]
    for word in vocab:
        fw.write(word[0]+'\n')
```

```python
vocab = list()
fr = open('vocab.txt', 'r', encoding='utf-8')
lines = fr.readlines()
for line in lines:
    vocab.append(line.rstrip())
# vocab.remove('\n')
vocab.remove(':')
vocab.remove('')
vocab.remove('')
vocab.insert(0, 'UNKNOWN')
vocab.insert(1, 'START')
vocab.insert(2, 'END')
vocab = vocab[:6000]
print(vocab)
```

```python
word2Index = {word: index for index, word in enumerate(vocab)}  #将词典映射成{词:0, 典:1}，类似这样
fr = open('Poetry.txt', 'r', encoding='utf-8')  #读取文件，准备数据集
lines = fr.readlines()  #所有行
dataSet = list()    #数据集
labels = list() #标签
for line in lines:
    try:
        poetry = line.split(":")[1].rstrip()  #去除标题
        X = [word2Index.get(word, 0) for word in list(poetry)] #把对应的文字转成索引，形成向量
        y = X.copy()
        X.insert(0, 1)  #1代表是词典里的START，代表开始
        dataSet.append(X)
        y.append(2) #2代表是词典里的END
        labels.append(y)
    except:
        pass
    print(dataSet[:100],labels[:100])
```
